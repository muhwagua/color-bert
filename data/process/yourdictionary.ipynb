```python
base_url = "https://sentence.yourdictionary.com/"
```

```python
colors = [
    "red", 
    "orange", 
    "yellow",
    "green",
    "blue",
    "purple",
    "brown",
    "white",
    "black",
    "pink",
    "lime",
    "gray",
    "violet",
    "cyan",
    "magenta",
    "khaki",
]
```

```python
import os
import requests
from bs4 import BeautifulSoup
```

```python
os.getcwd()
```

```python
os.mkdir("colors")
```

```python
%ls
```

```python
def scrape(color, save_dir="colors", return_sentences=False):
    url = f"{base_url}{color}"
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")
    divs = soup.find_all("div", class_="sentence-item")
    sentences = "\n".join(div.text for div in divs)
    with open(os.path.join(save_dir, f"{color}.txt"), "w+") as f:
        f.write(sentences)
    if return_sentences:
        return sentences
```

```python
red_sentences = scrape("red", return_sentences=True)
red_sentences
```

```python
for color in colors:
    scrape(color)
```
