```
def get_sentence(color):
    color_df = pd.DataFrame(columns = ('color', 'sentence'))
    idx = 0

    url = "https://sentencedict.com/"+color+"_1.html"
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    page_list = soup.select('#div_main_left > div')[1].text
    p = re.findall("\d+", page_list)
    page_num = p[3]

    for page in range(1, int(page_num)+1):
        url = "https://sentencedict.com/"+color+"_"+str(page)+".html"
        page = requests.get(url)
        soup = BeautifulSoup(page.content, 'html.parser')

        sentence_list = soup.select('div#all div')
    
        for s in sentence_list:
            sentence = s.get_text()
            if color in sentence.lower():
                color_df.loc[idx] = [color, sentence]
                idx += 1
    
    return color_df
```
